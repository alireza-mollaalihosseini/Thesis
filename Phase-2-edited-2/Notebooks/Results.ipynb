```python
import pandas as pd
import plotly.express as px
```

```python
df1 = pd.read_csv("/content/GridSearch_Results(Model-param) (1).csv")
df1
```

```python
fig = px.line(df1, x='M', y='avr_tref', color = 'lam', text='avr_tref')
fig.update_layout(title = { 'text' : "Avr_tref--vs--M\n (N_0= 1500, N0= 800)", 'x' : 0.5 })
fig.update_traces(textposition="middle center")
fig.show()
```

```python
fig = px.line(df1, x='M', y='avr_ref_time', color = 'lam', text='avr_ref_time')
fig.update_layout(title = { 'text' : "Average reference time --vs --M\n (N_0= 1500, N0= 800)", 'x' : 0.5 })
fig.update_traces(textposition="middle center")
fig.show()
```
